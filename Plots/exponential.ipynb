```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
k = 0.5  # You can adjust this to control how fast it decays

# Generate x values
x = np.linspace(0, 10, 1000)

# Define the exponential decay function
def exponential_decay(x, k=0.5):
    return 0.6 + 0.4 * np.exp(-k * x)

# Calculate y values
y = exponential_decay(x, k)

# To create a PDF, we need to normalize the function so it integrates to 1
# First, let's shift and scale to make it a proper probability density

# Since we want a PDF, we need y >= 0 and integral = 1
# Let's create a modified version that can serve as a PDF
def exponential_pdf(x, k=0.5):
    # Use the decay part only (0.4 * exp(-k*x)) and normalize
    return k * np.exp(-k * x)  # This is already normalized (integrates to 1)

# Generate PDF values
y_pdf = exponential_pdf(x, k)
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(6, 6))

# Plot the original exponential decay
ax1.plot(x, y, 'b-', linewidth=2)
ax1.set_xlabel('Group Size')
ax1.set_ylabel('Proportion of Samples with Mean > 0')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_xlim(0, 8)
ax1.set_ylim(0, 1)

# Remove x-axis tick marks and labels
ax1.set_xticks([])

plt.tight_layout()
plt.show()
```
